```python
import yfinance as yf

appl = yf.Ticker("AAPL")
```

```python
appl
```

```python
# get all stock info
appl.info
```

```python
appl.info['financialCurrency']
```

```python
# get historical market data
hist = appl.history(period="1mo")
```

```python
# show meta information about the history (requires history() to be called first)
appl.history_metadata
```

```python
# show actions (dividends, splits, capital gains)
appl.actions
appl.dividends
appl.splits
appl.capital_gains
```

```python
# show share count
appl.get_shares_full(start="2022-01-01", end=None)
```

```python
# show financials:
# - income statement\n
appl.income_stmt
appl.quarterly_income_stmt
```

```python
# - balance sheet
appl.balance_sheet
appl.quarterly_balance_sheet
```

```python
# - cash flow statement
appl.cashflow
appl.quarterly_cashflow
# see `Ticker.get_income_stmt()` for more options
```

```python
appl.get_income_stmt()
```

```python
# show holders
appl.major_holders
appl.institutional_holders
appl.mutualfund_holders
appl.insider_transactions
appl.insider_purchases
appl.insider_roster_holders
```

```python
# show recommendations
appl.recommendations
appl.recommendations_summary
appl.upgrades_downgrades
```

```python
# Show future and historic earnings dates, returns at most next 4 quarters and last 8 quarters by default. 
# Note: If more are needed use msft.get_earnings_dates(limit=XX) with increased limit argument.
appl.earnings_dates
```

```python
# show ISIN code - *experimental*
# ISIN = International Securities Identification Number
appl.isin
```

```python
# show options expirations
appl.options
```

```python
# show news
appl.news
```

```python
# get option chain for specific expiration
opt = appl.option_chain('2024-02-23')
opt.calls
# data available via: opt.calls, opt.puts
```
